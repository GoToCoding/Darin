```python
model = Sequential()
model.add(Conv2D(64, kernel_size=(5, 5), padding='same', activation='relu', input_shape=(15, 15, 3)))
model.add(BatchNormalization())
model.add(Conv2D(32, kernel_size=(5, 5), padding='same', activation='relu'))
model.add(BatchNormalization())
model.add(Conv2D(32, kernel_size=(5, 5), padding='same', activation='relu'))
model.add(BatchNormalization())
model.add(Dropout(0.5))
model.add(Flatten())
model.add(Dense(1024, activation='relu'))
model.add(Dense(512, activation='relu'))
model.add(BatchNormalization())
model.add(Dense(225, activation='softmax'))

model.compile(loss=keras.losses.categorical_crossentropy,
              optimizer=keras.optimizers.Adamax(),
              metrics=['accuracy'])

model.summary()
```

```python
def rotate_coordinates(x, y):
    return 14 - y, x


def rotate_randomly(table, ans_x, ans_y):
    k = random.randint(0, 3)
    res = np.zeros((15, 15, 3))
    for i in range(15):
        for j in range(15):
            x, y = rotate_coordinates(i, j)
            if k >= 1:
                x, y = rotate_coordinates(x, y)
            if k >= 2:
                x, y = rotate_coordinates(x, y)
            if k >= 3:
                x, y = rotate_coordinates(x, y)
            res[x][y][0] = table[i][j][0]
            res[x][y][1] = table[i][j][1]
            res[i][j][2] = table[0][0][2]
    ans_x, ans_y = rotate_coordinates(ans_x, ans_y)
    if k >= 1:
        ans_x, ans_y = rotate_coordinates(ans_x, ans_y)
    if k >= 2:
        ans_x, ans_y = rotate_coordinates(ans_x, ans_y)
    if k >= 3:
        ans_x, ans_y = rotate_coordinates(ans_x, ans_y)
    return res, ans_x, ans_y


def generator(raw_data, batch_size, data_max_len=-1):
    if data_max_len == -1:
        data_max_len = len(raw_data)

    def empty_table():
        x = np.zeros((15, 15, 3))
        for i in range(15):
            for j in range(15):
                x[i][j][2] = -1
        return x

    cur_game = 0
    cur_move = -1
    cur_player = -1
    cur_table = empty_table()

    while True:
        batch = [[], []]

        while len(batch[0]) != batch_size:
            if (cur_move == -1
                or raw_data[cur_game][0] != 'black'
                or cur_move + 2 >= len(raw_data[cur_game])
                or len(raw_data[cur_game]) < 6
            ):
                cur_game = (cur_game + 1) % data_max_len
                cur_move = 0
                cur_table = empty_table()
                continue

            game = raw_data[cur_game]
            if cur_move == 0:
                cur_move += 1
                x, y = translate_chess_coordinates(game[cur_move])
                if not go(cur_table, x, y, -1):
                    cur_move = -1
                    continue

                cur_move += 1
                x, y = translate_chess_coordinates(game[cur_move])
                if not go(cur_table, x, y, 1):
                    cur_move = -1
                    continue

            to_add = cur_table.copy()

            cur_move += 1
            x, y = translate_chess_coordinates(game[cur_move])
            if not go(cur_table, x, y, -1):
                cur_move = -1
                continue

            to_add, x, y = rotate_randomly(to_add, x, y)

            batch[0].append(to_add)
            batch[1].append(15 * x + y)
            cur_move += 1
            x, y = translate_chess_coordinates(game[cur_move])
            if not go(cur_table, x, y, 1):
                cur_move = -1
                continue

        yield np.array(batch[0]), keras.utils.to_categorical(batch[1], 225)


def go(table, x, y, color):
    if table[x][y][0] != 0 or table[x][y][1] != 0:
        return False
    if color == -1:
        table[x][y][0] = -1
    else:
        table[x][y][1] = 1
    return True


def translate_chess_coordinates(coord):
    x = ord(coord[0]) - ord('a')
    if x > ord('i') - ord('a'):
        x -= 1
    y = int(coord[1:]) - 1
    return x, y
```
